```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris, load_wine, load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
print("Original DataFrames:")
print(df.head())

scaler = StandardScaler()
df.iloc[:, :-1] = scaler.fit_transform(df.iloc[:, :-1])
print("\nDataFrame after scaling:")
print(df.head())

x_train,x_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df['target'], test_size=0.2, random_state=42)

print("\nshapes of train and test sets:")
print(f"x_train shape: {x_train.shape}, x_test shape: {x_test.shape}")
print(f"y_train shape: {y_train.shape}, y_test shape: {y_test.shape}")



from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix

models ={
    "Decision Tree" : DecisionTreeClassifier(),
    "Random forest" : RandomForestClassifier(),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier()
}

for name, model in models.items():
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)
  print(f"{name}:")
  print(confusion_matrix(y_test, y_pred))
  print(classification_report(y_test, y_pred))
```

```python
# supervised learning: Linerar regresision
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor # Corrected the import statement
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Train models
regressors = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(n_estimators=100) # Corrected the typo and the parameter name
}

for nme, model in regressors.items():
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)
  print(f"Model:{nme}") # Corrected the variable name
  print(f"MSE: {mean_squared_error(y_test, y_pred)}")
  print(f"R2: {r2_score(y_test, y_pred)}")
  print("\n")
```
